```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv('./open/train.csv')
test_df = pd.read_csv('./open/test.csv') 
```

```python
print(train_df.shape)
print(test_df.shape)
```

```python
# train
train_df['base_date'] = train_df['base_date'].astype(object)
train_df['base_hour'] = train_df['base_hour'].astype(object)

yyyy = train_df['base_date'].astype(str).str[0:4]
mm = train_df['base_date'].astype(str).str[4:6]
dd = train_df['base_date'].astype(str).str[6:8]
tt = train_df['base_hour'].astype(str) # 추가

train_df['base_date_obj'] = yyyy+'-'+mm+'-'+dd
train_df['base_hour_obj'] = yyyy+'-'+mm+'-'+dd + ' ' +tt+':00:00'

train_df['base_date_edited'] = pd.to_datetime(train_df['base_date_obj'])
train_df['base_hour_edited'] = pd.to_datetime(train_df['base_hour_obj'])
train_df.head(3)
```

```python
# test
test_df['base_date'] = test_df['base_date'].astype(object)
test_df['base_hour'] = test_df['base_hour'].astype(object)

yyyy = test_df['base_date'].astype(str).str[0:4]
mm = test_df['base_date'].astype(str).str[4:6]
dd = test_df['base_date'].astype(str).str[6:8]
tt = test_df['base_hour'].astype(str) # 추가

test_df['base_date_obj'] = yyyy+'-'+mm+'-'+dd
test_df['base_hour_obj'] = yyyy+'-'+mm+'-'+dd + ' ' +tt+':00:00'

test_df['base_date_edited'] = pd.to_datetime(test_df['base_date_obj'])
test_df['base_hour_edited'] = pd.to_datetime(test_df['base_hour_obj'])
test_df.head(3)
```

### 전체

```python
start = []
end = []
num_lst = []

for lat in unique:
    quary2 = train_df[(train_df.start_latitude ==lat)]
    num = quary2.end_latitude.nunique()
    
    slat = quary2.start_latitude.unique() # node_name
    elat = quary2.end_latitude.unique()

    start.append(slat)
    end.append(elat)
    num_lst.append(num)
        
table = pd.DataFrame(None)
table['start'] = start
table['end'] = end
table['number'] = num_lst
table
```

```python
table['number'].value_counts()
```

### number1

```python
table_1 = table[table['number']==1]
table_1 = table_1.reset_index(drop=True)
table_1

start = []
end = []

for i in range(len(table_1)):
    tmp = table_1['start'][i][0]
    tmp1 = table_1['end'][i][0]
    
    start.append(tmp)
    end.append(tmp1)
    
table_1_1 = pd.DataFrame(None)
table_1_1['start'] = start
table_1_1['end'] = end
table_1_1
```

### number2

```python
table_2 = table[table['number']==2]
table_2 = table_2.reset_index(drop=True)
table_2
```

```python
start = []
end = []

for i in range(len(table_2)):
    tmp = table_2['start'][i][0]
    tmp1 = table_2['end'][i][0]
    tmp2 = table_2['end'][i][1]
    
    start.append(tmp)
    start.append(tmp)
    end.append(tmp1)
    end.append(tmp2)
    
table_2_1 = pd.DataFrame(None)
table_2_1['start'] = start
table_2_1['end'] = end
table_2_1
```
